Question 1: What is the role of filters and feature maps in Convolutional Neural Network (CNN)?

Answer:

Filters, also known as kernels, are small matrices of learnable weights that scan across an input image to detect important features such as edges, corners, textures, and shapes. During the convolution operation, each filter extracts a specific pattern from different regions of the image. The output produced after applying a filter is called a feature map (or activation map), which highlights the locations where that particular feature is detected. As the network becomes deeper, filters learn increasingly complex features, such as object parts and complete objects. Thus, filters are responsible for feature extraction, while feature maps store the extracted information that is passed to the next layers of the CNN.
Question 2: Explain the concepts of padding and stride in CNNs (Convolutional Neural Network). How do they affect the output dimensions of feature maps?

Answer:

Padding is the process of adding extra rows and columns, usually filled with zeros, around the borders of an input image before applying convolution. It helps preserve spatial dimensions and ensures that edge pixels are considered during feature extraction. Stride refers to the number of pixels by which the filter moves across the image during convolution. A stride of 1 moves the filter one pixel at a time, producing larger feature maps, while larger stride values reduce the size of the output feature map and decrease computational cost. Therefore, padding generally increases or preserves the output dimensions, whereas increasing the stride decreases the dimensions of the resulting feature maps.
Question 3: Define receptive field in the context of CNNs. Why is it important for deep architectures?

Answer:

The receptive field in a Convolutional Neural Network is the region of the input image that influences the activation of a particular neuron in a feature map. In the initial layers, the receptive field is small and captures simple local features such as edges and textures. As more convolutional and pooling layers are added, the receptive field grows larger, allowing neurons in deeper layers to capture broader contextual information and more complex patterns. A larger receptive field is essential in deep architectures because it enables the network to recognize large objects, understand spatial relationships, and improve overall classification accuracy by combining both local and global image features.
Question 4: Discuss how filter size and stride influence the number of parameters in a CNN.

Answer:

The filter size determines the number of learnable parameters in a convolutional layer. Larger filters contain more weights, increasing both the model's complexity and computational cost, whereas smaller filters require fewer parameters and are more computationally efficient. The number of parameters in a convolutional layer depends on the filter dimensions, input channels, and the number of filters, but it is independent of the input image size. Stride does not affect the number of trainable parameters because it only determines how far the filter moves across the input image. However, a larger stride reduces the size of the output feature maps, decreases computation, and lowers memory usage by producing fewer activations.
Question 5: Compare and contrast different CNN-based architectures like LeNet, AlexNet, and VGG in terms of depth, filter sizes, and performance.

Answer:

LeNet, AlexNet, and VGG are landmark CNN architectures that demonstrate the evolution of deep learning models. LeNet, introduced in 1998, is a relatively shallow network with 5 learnable layers and was designed primarily for handwritten digit recognition using small convolution filters and average pooling. AlexNet, introduced in 2012, contains 8 learnable layers and uses larger filters in the first layer along with ReLU activation, dropout, and GPU training, achieving a significant improvement in image classification accuracy on the ImageNet dataset. VGG, introduced in 2014, further increased the network depth to 16 or 19 layers while consistently using small 3×3 convolution filters throughout the architecture. Although VGG achieves higher accuracy and learns more detailed features than LeNet and AlexNet, it requires significantly more parameters, computational power, and memory. Overall, LeNet is simple and lightweight, AlexNet introduced several innovations that advanced deep learning, and VGG demonstrated that increasing depth with smaller filters can substantially improve performance on complex image recognition tasks.

```python
# Question 6:
# Using Keras, build and train a simple CNN model on the MNIST dataset
# from scratch. Include code for model creation, compilation,
# training, and evaluation.

import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load Dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# Reshape
X_train = X_train.reshape(-1, 28, 28, 1)
X_test = X_test.reshape(-1, 28, 28, 1)

# CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),

    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train
history = model.fit(
    X_train,
    y_train,
    epochs=3,
    validation_split=0.2
)

# Evaluate
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", accuracy)
```

The CNN model was successfully built using Keras with two convolutional layers, two max-pooling layers, and two fully connected layers. After training on the MNIST dataset for three epochs, the model achieved high classification accuracy on the test dataset, demonstrating the effectiveness of CNNs for handwritten digit recognition.

```python
# Question 7:
# Load and preprocess the CIFAR-10 dataset using Keras,
# and create a CNN model to classify RGB images.
# Show your preprocessing and architecture.

import tensorflow as tf
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load Dataset
(X_train, y_train), (X_test, y_test) = cifar10.load_data()

# Normalize
X_train = X_train / 255.0
X_test = X_test / 255.0

# CNN Model
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    MaxPooling2D((2,2)),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D((2,2)),

    Conv2D(128, (3,3), activation='relu'),

    Flatten(),

    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

history = model.fit(
    X_train,
    y_train,
    epochs=3,
    validation_split=0.2
)

loss, accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", accuracy)
```

The CIFAR-10 dataset was normalized by scaling pixel values between 0 and 1 before training. A CNN containing convolutional layers, max-pooling layers, and fully connected layers was created to classify RGB images into ten categories. After training, the model learned useful image features and produced satisfactory classification accuracy on the test dataset.

```python
# Question 8:
# Using PyTorch, write a script to define and train a CNN
# on the MNIST dataset.
# Include model definition, data loaders,
# training loop, and accuracy evaluation.

import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Dataset
transform = transforms.ToTensor()

train_dataset = datasets.MNIST(
    root='./data',
    train=True,
    download=True,
    transform=transform
)

test_dataset = datasets.MNIST(
    root='./data',
    train=False,
    download=True,
    transform=transform
)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=64)

# CNN Model
class CNN(nn.Module):

    def __init__(self):
        super().__init__()

        self.conv = nn.Sequential(
            nn.Conv2d(1,32,3),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(32,64,3),
            nn.ReLU(),
            nn.MaxPool2d(2)
        )

        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(1600,128),
            nn.ReLU(),
            nn.Linear(128,10)
        )

    def forward(self,x):
        x=self.conv(x)
        x=self.fc(x)
        return x

model = CNN()

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training
for epoch in range(3):

    model.train()

    for images, labels in train_loader:

        optimizer.zero_grad()

        outputs = model(images)

        loss = criterion(outputs, labels)

        loss.backward()

        optimizer.step()

    print(f"Epoch {epoch+1} completed")

# Evaluation
correct = 0
total = 0

model.eval()

with torch.no_grad():

    for images, labels in test_loader:

        outputs = model(images)

        _, predicted = torch.max(outputs,1)

        total += labels.size(0)

        correct += (predicted == labels).sum().item()

print("Accuracy:",100*correct/total,"%")
```

A CNN model was implemented using PyTorch with convolutional, ReLU, max-pooling, and fully connected layers. The model was trained on the MNIST dataset using the Adam optimizer and Cross-Entropy Loss. After training, the model achieved high accuracy on the test dataset, demonstrating successful handwritten digit classification.

```python
# Question 9:
# Given a custom image dataset stored in a local directory,
# write code using Keras ImageDataGenerator to preprocess
# and train a CNN model.

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Dataset Directory
train_path = "dataset/train"
validation_path = "dataset/validation"

# Data Generator
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='binary'
)

# CNN Model
model = Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape=(128,128,3)),
    MaxPooling2D(2),

    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D(2),

    Flatten(),

    Dense(128,activation='relu'),

    Dense(1,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=3
)
```

The custom image dataset was preprocessed using the ImageDataGenerator class, which performed normalization and data augmentation to improve model generalization. A CNN model was then trained on the processed images using binary cross-entropy loss and the Adam optimizer. This workflow helps improve classification accuracy while reducing overfitting.
